```python
import requests
import json
from datetime import datetime
import pandas as pd

# Cabeceras necesarias para la consulta de la API
headers = {
    "Accept": "application/json; application/vnd.esios-api-v1+json",
    "Content-Type": "application/json",
    "x-api-key": "token-personal"
}

# Parametros de la consulta
params = {
    "start_date": "2021-01-01T00:00:00.000Z",
    "end_date": "2022-12-31T23:59:59.000Z",
    "time_trunc": "hour",
    "time_agg": "avg",
    
}

# Realizar consulta con la libreria request
response = requests.get("https://api.esios.ree.es/indicators/551", headers=headers, params=params)

# Comprobaciones
if response.status_code == 200:
    data = json.loads(response.text)
    values = data["indicator"]["values"]

else:
    print(f"Error: {response.status_code}")
    
df_data = pd.DataFrame(values)
df_data.to_csv('eolica_2021-2022_API.csv')
```
